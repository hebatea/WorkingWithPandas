```python
# read Execl Files
def convert_event_cell(cell):
    if cell =='n.a.':
        return 'Sunny'
    return cell

def convert_temperature_cell(cell):
    if cell == 'n.a.':
        return None
    return cell
df = pd.read_excel('C:\\Users\\Sunshine\\Desktop\\Data Analysis\\weather_data.xlsx', 'Sheet1', converters ={
    'event': convert_event_cell,
    'temperature': convert_temperature_cell
})
df
```

```python
# write to excel
df.to_excel('C:\\Users\\Sunshine\\Desktop\\Data Analysis\\new.xlsx', sheet_name='weather', 
            startrow=1, startcol= 1, index = False)
```
