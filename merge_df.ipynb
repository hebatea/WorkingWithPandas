```python
import pandas as pd
```

```python
df_temperature = pd.DataFrame({
    'city': ['New York', 'chicago', 'Texas', 'New Jercy'],
    'temperature': [21, 42, 35, 70]
})
df_temperature
```

```python
df_humidity = pd.DataFrame({
    'city': ['New York', 'chicago', 'Las Vags'],
    'humidity': [30, 20, 40]
})
df_humidity
```

```python
#here it will merge common city btwn two dataframes (inner join)
df_merge = pd.merge(df_temperature, df_humidity, on= 'city')
df_merge
```

```python
#(outer join)
df_merge = pd.merge(df_temperature, df_humidity, on= 'city', how= 'outer')
df_merge
```

```python
#(inner join) -> by default
df_merge = pd.merge(df_temperature, df_humidity, on= 'city', how= 'inner')
df_merge
```

```python
#left join -> all df_temperature and common btwn df_temperature, df_humidity
df_merge = pd.merge(df_temperature, df_humidity, on= 'city', how= 'left')
df_merge
```

```python
#right join -> all df_humidity and common btwn df_temperature, df_humidity
df_merge = pd.merge(df_temperature, df_humidity, on= 'city', how= 'right')
df_merge
```

```python
#indicator -> to know from which data frame values are
df_merge = pd.merge(df_temperature, df_humidity, on= 'city', how= 'outer', indicator= True)
df_merge
```

```python
df1 = pd.DataFrame({
    'city': ['New York', 'chicago', 'Texas', 'New Jercy'],
    'temperature': [21, 42, 35, 70],
    'humidity': [32, 25, 45, 44]
})
df1
```

```python
df2 = pd.DataFrame({
    'city': ['New York', 'chicago', 'Las Vags'],
    'temperature': [21, 42, 30],
    'humidity': [32, 25, 40]
})
df2
```

```python
df3 = pd.merge(df1, df2, on= 'city')
df3
```

```python
df4 = pd.merge(df1, df2, on= 'city', suffixes= ('_left', '_right'))
df4
```
